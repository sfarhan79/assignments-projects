```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
import statsmodels.api as sm
from statsmodels.graphics.regressionplots import influence_plot
```

```python
fifty=pd.read_csv("https://raw.githubusercontent.com/sfarhan79/datasets/main/python%20datasets/csv/fifty%20start-ups.csv")
fifty.head()
```

```python
fifty.info()
```

```python
fif=fifty.rename({"R&D Spend":"RDS","Administration":"ADMS","Marketing Spend":"MKS"},axis=1)
fif.head()
```

```python
#Building Model
model=smf.ols("Profit~RDS+ADMS+MKS",data=fif).fit()
```

```python
#building SLR and MLR models for insignificant variables ADMS and MKS. Finding tvalues and pvalues
slr_a=smf.ols("Profit~ADMS",data=fif).fit()
slr_a.tvalues,slr_a.pvalues #insignificant pvalue
```

```python
slr_m=smf.ols("Profit~MKS",data=fif).fit()
slr_m.tvalues,slr_m.pvalues #significant pvalue
```

```python
model_mlr=smf.ols("Profit~ADMS+MKS",data=fif).fit()
model_mlr.tvalues,model_mlr.pvalues #variables has significant pvalues
```

```python
#Model Validation
#Collinearity check (VIF)
rsq_r=smf.ols("RDS~ADMS+MKS",data=fif).fit().rsquared
vif_r=1/(1-rsq_r)

rsq_m=smf.ols("MKS~ADMS+RDS",data=fif).fit().rsquared
vif_m=1/(1-rsq_m)

rsq_a=smf.ols("ADMS~RDS+MKS",data=fif).fit().rsquared
vif_a=1/(1-rsq_a)

#putting values into DataFrame
df1={"Variables":["RDS","MKS","ADMS"],"VIF":[rsq_r,rsq_m,rsq_a]}
df=pd.DataFrame(df1)
df
```

**as no VIF value>20, thus no Collinearity**

```python
#Residual Analysis - Test for Normality of residuals (QQPlot) by residual model
sm.qqplot(model.resid,line="q")
plt.title("QQPLot for Normality of residuals")
plt.show()
```

```python
list(np.where(model.resid<-30000))
```

```python
#Test of Homoscedascity/Heteroscedascity (plotting model's fitted values vs residual values)
def standard_values(vals):
  return (vals-vals.mean())/vals.std()
```

```python
plt.scatter(standard_values(model.fittedvalues),standard_values(model.resid))
plt.title("Residual Plot")
plt.xlabel("Standardized fitted values")
plt.ylabel("Standardized residual values")
plt.show()
```

```python
#Test for errors using Regression plots
fig=plt.figure(figsize=(10,5))
sm.graphics.plot_regress_exog(model,"RDS",fig=fig)
plt.show()
```

```python
fig=plt.figure(figsize=(10,5))
sm.graphics.plot_regress_exog(model,"ADMS",fig=fig)
plt.show()
```

```python
fig=plt.figure(figsize=(10,5))
sm.graphics.plot_regress_exog(model,"MKS",fig=fig)
plt.show()
```

```python
#Checking outliers/influencers - Two Methods
#Cook's Distance
(c,_)=model.get_influence().cooks_distance
c
```

```python
fig=plt.figure(figsize=(10,5))
plt.stem(np.arange(len(fif)),np.round(c,5))
plt.xlabel("Row Index")
plt.ylabel("Cooks Distance")
plt.show()
```

```python
#Index and value of influencer where C>0.5
np.argmax(c),np.max(c)
```

```python
#Leverage values using high influence points. POints beyond leverage cutoff value are influencers
influence_plot(model)
plt.show()
```

```python
#Leverage cut-off value - 3*(k+1)/n
k=fif.shape[1]
n=fif.shape[0]
leverage_cutoff=3*(k+1)/n
leverage_cutoff
```

```python
fif[fif.index.isin([49])]
```

```python
#Improving the model
fiff=fif.drop(fif.index[[49]],axis=0).reset_index(drop=True)
fiff
```

```python
#Final Model
while np.max(c)>0.5:
  model=smf.ols("Profit~ADMS+RDS+MKS",data=fiff).fit()
  (c,_)=model.get_influence().cooks_distance
  c
  np.argmax(c), np.max(c)
  fiff=fiff.drop(fiff.index[[np.argmax(c)]],axis=0).resetindex(drop=True)
  fiff
else:
  final_model=smf.ols("Profit~ADMS+RDS+MKS",data=fiff).fit()
  final_model.rsquared,final_model.aic
  print("The model accuracy is improved to ",final_model.rsquared)
```
