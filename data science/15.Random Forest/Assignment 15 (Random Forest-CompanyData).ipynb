```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
companyd=pd.read_csv("https://raw.githubusercontent.com/sfarhan79/datasets/main/python%20datasets/csv/company%20data.csv")
companyd.head()
```

```python
companyd.info()
```

```python
companyd.shape
```

```python
fig=plt.figure(figsize=(4,4))
sns.pairplot(data=companyd,hue="ShelveLoc")
```

```python
cdata=pd.get_dummies(companyd,columns=["Urban","US"],drop_first=True)
cdata.head()
```

```python
cdata["ShelveLoc"]=cdata["ShelveLoc"].map({"Good":1,"Medium":2,"Bad":3})
cdata.head()
```

```python
X=cdata.iloc[:,0:6]
Y=cdata["ShelveLoc"]
```

```python
labels=np.array(cdata["Income"])
features=cdata.drop("Income",axis=1)
features=np.array(cdata)
feature_list=list(cdata.columns)
```

```python
train_features,test_features,train_labels,test_labels=train_test_split(features,labels,test_size=0.20,random_state=40)
```

```python
#Establishing Baseline
baseline_pred=test_features[:,feature_list.index("Sales")]
baseline_error=abs(baseline_pred-test_labels)
print("Average Baseline Error:",round(np.mean(baseline_error),2))
```

```python
rf=RandomForestRegressor(n_estimators=1000,random_state=30)
rf.fit(train_features,train_labels)
```

```python
preds=rf.predict(test_features)
errors=abs(preds-test_labels)
print("Mean Absolute Error:",round(np.mean(errors),2),"degrees")
```

```python
#Mean Absolute Percentage Error (MAPE)
mape=100*(errors/test_labels)
accuracy=100-np.mean(mape)
print("Mean Absolute Percentage Error:",round(np.mean(accuracy)),"%")
```
