```python
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine
```

```python
books=pd.read_csv("https://raw.githubusercontent.com/sfarhan79/datasets/main/python%20datasets/csv/book_2.csv",encoding="Latin1")
books.head()
```

```python
books=books.rename({"User.ID":"user_ID","Book.Title":"book_title","Book.Rating":"rating"},axis=1)
books.head()
```

```python
books.shape
```

```python
books.info()
```

```python
book=books.iloc[:,1:]
book.head()
```

```python
len(book.user_ID.unique())
```

```python
len(book.book_title.unique())
```

```python
bk=book.pivot_table(index="user_ID",columns="book_title",values="rating").reset_index(drop=True)
bk
```

```python
bk.index=book.user_ID.unique()
bk
```

```python
bk.fillna(0,inplace=True)
bk
```

```python
bk_pd=1-pairwise_distances(bk,metric="cosine")
bk_pd
```

```python
bk_pdf=pd.DataFrame(bk_pd)
bk_pdf
```

```python
bk_pdf.index=book.user_ID.unique()
bk_pdf.columns=book.user_ID.unique()
bk_pdf
```

```python
np.fill_diagonal(bk_pd,0)
bk_pdf
```

```python
bk_pdf.idxmax(axis=1)
```
