```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
wine=pd.read_csv("https://raw.githubusercontent.com/sfarhan79/datasets/main/python%20datasets/csv/wine.csv")
wine.head()
```

```python
wine.info()
```

```python
wine.Type.value_counts()
```

```python
wine2=wine.iloc[:,1:]
wine2
```

```python
wine_array=wine2.values
wine_array
```

```python
from sklearn.preprocessing import scale
wine_norm=scale(wine_array)
wine_norm
```

```python
pa=PCA(n_components=13)
paf=pa.fit_transform(wine_norm)
paf
```

```python
pa.components_
```

```python
#The amount of ratio each PCA has.
var=pa.explained_variance_ratio_
var
```

```python
#Cumulative variance of each PCA
var1=np.cumsum(np.round(var,4)*100)
var1
```

```python
plt.plot(var1,color="magenta")
```

```python
final_df=pd.concat([wine.Type,pd.DataFrame(paf[:,0:3],columns=["PCA1","PCA2","PCA3"])],axis=1)
final_df
```

```python
fig=plt.figure(figsize=(10,5))
sns.scatterplot(data=final_df)
```

```python
#KMeans Clustering
cc=[]
for i in range(1,6):
  km=KMeans(n_clusters=i,random_state=35)
  kmf=km.fit(wine_norm)
  cc.append(kmf.inertia_)

plt.plot(range(1,6),cc)
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("CC")
plt.legend()
plt.show()
```

```python
kl=KMeans(3,random_state=54).fit(wine_norm)
kl
```

```python
set(kl.labels_)
```

```python
wine3=wine.copy()
wine3["cluster_ID3"]=kl.labels_
wine3
```

```python
wine3.cluster_ID3.value_counts()
```

```python
#Heirarchical Clustering
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
from sklearn.preprocessing import normalize
```

```python
plt.figure(figsize=(10,5))
deng=sch.dendrogram(sch.linkage(wine_norm,"complete"))
```

```python
agc=AgglomerativeClustering(n_clusters=3,affinity="euclidean",linkage="ward")
agc
```

```python
y=pd.DataFrame(agc.fit_predict(wine_norm),columns=["clusters_ID"])
y["clusters_ID"].value_counts()
```

```python
wine4=wine.copy()
wine4["cluster_ID4"]=agc.labels_
wine4
```
