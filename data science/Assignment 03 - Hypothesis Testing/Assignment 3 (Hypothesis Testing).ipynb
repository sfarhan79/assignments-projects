**Q1**

```python
#Null H: no difference in diameter of cutlets
#Alt. H:there is a difference in diameter of cutlets.
import pandas as pd
import numpy as np
from scipy import stats
```

```python
cutlet=pd.read_csv("https://raw.githubusercontent.com/sfarhan7/datasets/main/python%20datasets/csv/cutlets.csv")
cutlet.head()
```

```python
cutletA=pd.Series(cutlet.iloc[:,0])
cutletA.head()
```

```python
cutletB=pd.Series(cutlet.iloc[:,1])
cutletB.head()
```

```python
cutlet2=stats.ttest_ind(cutletA,cutletB)
cutlet2
```

Conclusion:  p>alpha(0.05) Hence, Null H should be accepted, which says there is no difference in diameter of cutlets.

**Q2**

```python
#Null H: no difference in TAT
#Alt. H: there is difference in TAT
tat=pd.read_csv("https://raw.githubusercontent.com/sfarhan7/datasets/main/python%20datasets/csv/labTAT.csv")
tat.head()
```

```python
tat2=stats.f_oneway(tat.iloc[:,0],tat.iloc[:,1],tat.iloc[:,2],tat.iloc[:,3])
tat2
```

Conclusion: p>alpha(0.05) Hence Null H should be accepted. There is no difference in TAT.

**Q3**

```python
#Null H:M-F buyer ratio is similar across regions
#Alt. H:M-F buyer ratio is different across regions
sales=pd.read_csv("https://raw.githubusercontent.com/sfarhan7/datasets/main/python%20datasets/csv/buyerratio.csv")
sales
```

```python
salesm=[50,142,131,70]
salesf=[435,1523,1356,750]
sales2=[salesm,salesf]
sales2
```

```python
from scipy.stats import chi2_contingency
print(chi2_contingency(sales2))
```

Conclusion: p>0.05 Accept Null H.
